## 自定义坐标轴数字，及对应的数字替换

```python
import matplotlib.pyplot as plt
import numpy as np
```

准备数据

```python
X = np.linspace(-np.pi, np.pi, 300)

C = np.cos(X)
S = np.sin(X)
```

绘图

```python
plt.plot(X, C)
plt.plot(X, S)

plt.xticks([-3, -2, -1, 0, 1, 2, 3])            # 定义坐标轴要所显示的数字，可以没有顺序
plt.yticks([-1, 0, 1], ['A', 'B', 'C'])         # 将对应的数字替换为其自定义的含义

plt.show()
```
