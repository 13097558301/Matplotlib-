## 显示图例

```python
import matplotlib.pyplot as plt
import numpy as np
```

准备数据

```python
plt.rcParams['font.sans-serif'] = ['SimHei']    #  显示中文的设置
plt.rcParams['axes.unicode_minus'] = False      #  显示负号的设置 

X = np.linspace(-np.pi, np.pi, 300)

C = np.cos(X)
S = np.sin(X)
```

绘图

```python
plt.plot(X, C, label='cos()')                 # 设置线段的名称
plt.plot(X, S, label='sin()')

plt.legend(loc='lower right', title='图例')   # 让图例生效

plt.show()
```

### 关于设置图例的相关参数

* 设置图例位置：使用loc参数 ，  plt.legend(loc='lower left')

    * 0: 'best'
    * 1: 'upper right'
    * 2: 'upper left'
    * 3: 'lower left'
    * 4: 'lower right'
    * 5: 'right'
    * 6: 'center left'
    * 7: 'center right'
    * 8: 'lower center'
    * 9: 'upper center'
    * 10: 'center'

* 设置图例字体:设置字体大小
    * fontsize : int or float or {‘xx-small’, ‘x-small’, ‘small’, ‘medium’, ‘large’, ‘x-large’, ‘xx-large’}
        

        

* 设置图例边框及背景
    * plt.legend(loc='best',frameon=False) 	#去掉图例边框

    * plt.legend(loc='best',edgecolor='blue')     #设置图例边框颜色

    * plt.legend(loc='best',facecolor='blue')      #设置图例背景颜色,若无边框,参数无效

* 设置图例标题:

    * plt.legend(loc='best',title='figure 1 legend')   #图例标题
