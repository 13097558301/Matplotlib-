## 设置 图像 、x轴、y轴 标题

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']    #  显示中文的设置
plt.rcParams['axes.unicode_minus'] = False      #  显示负号的设置 
```

准备数据

```python
X = np.linspace(-np.pi, np.pi, 300)

C = np.cos(X)
S = np.sin(X)
```

设置 图像 、x轴、y轴 标题

```python
plt.plot(X, C)
plt.plot(X, S)

plt.title('正弦 、 余弦 函数图像')              # 显示图像标题
plt.xlabel('x 轴')                              # 显示图像 x轴 标题
plt.ylabel('y 轴')                              # 显示图像 y轴 标题

plt.show()
```
