对于图中图的设计方式，也可以转换为多个子图的设计。从某种角度上来说，add_axes() 所给定的权限也更高，当然，带来的问题就是需要自己进行手动调试的部分也就更多。

```python
import matplotlib.pyplot as plt
import numpy as np
```

数据准备

```python
X = np.linspace(-np.pi, np.pi, 300)

C = np.cos(X)
S = np.sin(X)
```

绘图

```python
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8      # 所代表的是占比
```

```python
fig = plt.figure()

ax_1 = fig.add_axes([left, bottom, width, height])

ax_1.plot(X, C)
ax_1.plot(X, S)

ax_2 = fig.add_axes([0.15, 0.6, 0.2, 0.22])
ax_2.plot(X, C)
ax_2.plot(X, S)

plt.show()
```
