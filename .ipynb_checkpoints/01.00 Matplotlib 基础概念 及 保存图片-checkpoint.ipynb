## Matplotlib 架构

对于使用 Matplotlib 进行画图，是有一个基础流程的，但是由于网上的资料过于庞杂，有些没有讲清楚，但是作为一个有着强迫症的程序员来说，有些事情必须捋清楚。

* figure 相当于一个画板；具体来说就是程序运行时跳出来的窗口。当然，你可以在一个程序中创建多个画板。
* axis 可以理解为一个画纸；一个 figure 上可以有一个或多个 axis 。
* matplotlib.pyplot 可以理解为画笔；例如：plt.plot 可以画连线图， plt.scatter 可以用来画散点图

从上述 画板、画纸、画笔 的关系，其实就可以大概看出 Matplotlib 的整体设计结构。

但是，话说回来，对于只有一张图的时候，并不会进行 figure,axis 的创建，往往都是plt.plot()直接画，因为这样效率高。

## Matplotlib，pyplot和pylab 它们之间有什么关系？

* Matplotlib 就是我们要使用的第三方包；
* matplotlib.pyplot 是 matplotlib 中的一个模块，用于绘图；
* pylab 是一个与 matplotlib 一起安装的模块，他谈不上是一个单独的包，因为它只是一个便利模块，它在单个名称空间中批量导入 matplotlib.pyplot 和 numpy。但是由于命名空间污染而不推荐用它，但也有很多大佬在边缘试探。个人也不喜欢用它，因为我认真且怂。

## 保存图片

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 300)

C = np.cos(X)
S = np.sin(X)

plt.plot(X, C)
plt.plot(X, S)

# 保存图片
plt.savefig("Cos&Sin.png",       # 填写保存地址及图片名称
            transparent=True,    # 图片背景是否透明
            format='png')        # 图片格式 

plt.show()
```
