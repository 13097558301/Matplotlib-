## 从简至繁：
先在一张图上绘制 正弦 、余弦 图像，然后再进行逐步美化

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(-np.pi, np.pi, 300)
```

X 是一个 numpy 数组，包含了从 −π 到 +π 等间隔的 300 个值。

```python
C = np.cos(X)
S = np.sin(X)
```

C 和 S 则分别是这 300 个值对应的余弦和正弦函数值组成的 numpy 数组。

```python
plt.plot(X, C)
plt.plot(X, S)

plt.show()
```

## 提前准备好一些数据

```python
x_list = np.linspace(-100, 100, 10)
y_list = np.linspace(-100, 100, 10)
x_list[9] = 0
y_list[9] = 50
```

## plt.plot()  基本属性----连线图

plt.plot() 函数的主要功能就是根据所给定的数据将图画出来, 然后根据点在数组中的顺序，plt.plot() 会自动将点连起来。

```python
plt.plot(x_list,          #  x 轴坐标
         y_list,          #  y 轴坐标
         marker = '<',    #  点型 
         color = 'b',     #  颜色
         linestyle = '-', #  线型
         linewidth = 1,   #  线段粗细
         label = 'L_1'    #  线段名称，绘制图例时要用
        )
plt.show()
```

x_list 和 y_list 这两个信息必填，且个数必须相等；其余均可不填。

* 有关点型（marker）的种类：
    * 像素： ',' （默认）
    * 圆形： 'o'
    * 上三角： '^'
    * 下三角： 'v'
    * 左三角： '<'
    * 右三角： '>'
    * 方形： 's'
    * 加号： '+'
    * 叉形： 'x'
    * 棱形： 'D'
    * 细棱形： 'd'
    * 三脚架朝下： '1'（像'丫'）
    * 三脚架朝上： '2'
    * 三脚架朝左： '3'
    * 三脚架朝右： '4'
    * 六角形： 'h'
    * 旋转六角形： 'H'
    * 五角形： 'p'
    * 垂直线： '|'
    * 水平线： '_'

* 有关颜色（color 简写为 c）的种类：
    * 蓝色： 'b' (blue) （默认）
    * 绿色： 'g' (green)
    * 红色： 'r' (red)
    * 蓝绿色(墨绿色)： 'c' (cyan)
    * 红紫色(洋红)： 'm' (magenta)
    * 黄色： 'y' (yellow)
    * 黑色： 'k' (black)
    * 白色： 'w' (white)

* 有关线型（linestyle 简写为 ls）的种类：
    * 实线： '-'  （默认）
    * 虚线： '--'
    * 虚点线： '-.'
    * 点线： ':'
    * 点： '.'
