## 等高线的绘制：plt.contour()

```python
import matplotlib.pyplot as plt
import numpy as np
```

准备数据

```python
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)

X, Y = np.meshgrid(x, y)
```

自定义函数f(x,y) 在这里只是一种计算登高线高度的公式，不必纠结

```python
def f(x, y):  
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)
```

绘图

```python
plt.contourf(X, Y, f(X, Y), 10, alpha=0.75, cmap=plt.cm.hot)  # 填充出颜色（10块区域）

C = plt.contour(X, Y, f(X, Y), 10, colors='k')                # 显示界限

plt.clabel(C, inline=True, fontsize=10)                       # 显示每条线所代表的值

plt.xticks(())
plt.yticks(())
plt.show()
```
