## x轴、y轴 的范围限制显示

```python
import matplotlib.pyplot as plt
import numpy as np
```

准备数据

```python
X = np.linspace(-np.pi, np.pi, 300)

C = np.cos(X)
S = np.sin(X)
```

x轴、y轴 的范围限定

```python
plt.plot(X, C)
plt.plot(X, S)

plt.xlim(-1, 1)      # 限制 x 轴显示范围
plt.ylim(-1, 1)      # 限制 y 轴显示范围

plt.show()
```
