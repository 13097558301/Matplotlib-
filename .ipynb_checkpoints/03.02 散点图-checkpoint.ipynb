使用 plt.scatter() 绘制散点图

```python
import matplotlib.pyplot as plt
import numpy as np
```

数据准备

```python
n = 1024
arr_x = np.random.normal(0, 1, n)
arr_y = np.random.normal(0, 1, n)
arr_color = np.arctan2(arr_y, arr_x)
```

绘图，一些常规参数

```python
plt.scatter(arr_x, arr_y, s=25, c=arr_color, alpha=0.5)
plt.show()
```
