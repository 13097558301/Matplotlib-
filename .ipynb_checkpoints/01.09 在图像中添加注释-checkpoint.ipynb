## 在图像中添加注释

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']    #  显示中文的设置
plt.rcParams['axes.unicode_minus'] = False      #  显示负号的设置 
```

准备数据

```python
X = np.linspace(-np.pi, np.pi, 300)

C = np.cos(X)
S = np.sin(X)
```

绘图 及 添加注释

* 方法一(有箭头）：plt.annotate('注释',xy=(1,2),xytext=(10,10),arrowprops = dict(facecolor = "k", headlength = 5, headwidth = 5, width = 1))
    * xy=(箭头所要指向的位置）
    * xytext=(注释所在的位置)
    * arrowprops： 箭头的一些参数，已调好，可以直接用

```python
plt.plot(X, C)
plt.plot(X, S)

plt.annotate('注释',
             xy=(0.5, 0.5),
             xytext=(1, 1),
             arrowprops = dict(facecolor = "k",
                               headlength = 5, 
                               headwidth = 5, 
                               width = 0.2)
            )

plt.show()
```

* 方法二（无箭头）： plt.text(x,y,'注释内容',fontdict={'size':16,'color':'k'})

```python
plt.plot(X, C)
plt.plot(X, S)

plt.text(0.5, 0.5, "注释", fontdict={'size':16,'color':'k'})

plt.show()
```
