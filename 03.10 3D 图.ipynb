在 jupyter notebook 中的 3D 图并不能进行全方位的观察，建议使用窗口进行观察。

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 这个包貌似在 matplotlib 内
```

准备数据

```python
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)                  # 生成坐标矩阵
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R)
```

绘图

```python
fig = plt.figure()
ax = Axes3D(fig)                         # 显示为 3D 坐标系


# 绘制3D图像
ax.plot_surface(X, Y, Z,                      # 数据
                rstride=1,                    # 指定行的跨度
                cstride=1,                    # 指定列的跨度
                cmap=plt.get_cmap('rainbow')  # 设计颜色的映射
               )    


ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-2, 2)

plt.show()
```
