```python
import matplotlib.pyplot as plt
import numpy as np
```

准备数据

```python
X = np.linspace(-3*np.pi, 3*np.pi, 300)

C = np.cos(X)
S = np.sin(X)
```

## 默认填充：函数会尽量凸显出图像之间的不同

```python
plt.plot(X, C)
plt.plot(X, S)

plt.fill(X, C)
plt.fill(X, S)

plt.show()
```

```python
plt.plot(X, C)
plt.plot(X, S)

plt.fill(X, C, X, S)

plt.show()
```

##  条件填充：

```python
plt.plot(X, C)
plt.plot(X, S)

plt.fill_between(x=X,           # 定义曲线的 x 坐标
                y1=C,           # 定义第一条曲线的 y 坐标
                y2=S,           # 定义第二条曲线的 y 坐标
                where=(C-S)>0,  # 条件表达式
                color='b',      # 填充颜色
                alpha=0.5)      # 透明度

plt.fill_between(x=X,
                y1=C,
                y2=S,
                where=(C-S)<0,
                color='r',
                alpha=0.5)

plt.show()
```
