```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']    #  显示中文的设置
plt.rcParams['axes.unicode_minus'] = False      #  显示负号的设置 
```

准备数据

```python
X = np.linspace(-np.pi, np.pi, 300)

C = np.cos(X)
S = np.sin(X)
```

## 创建多个窗口（图像） 及 在子图中实现常用操作

在 jupyter notebook 可能显示的并不是很明显，可以复制之后转到 PyCharm 运行显示。

```python
# 创建 画板-1 - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
fig_1 = plt.figure("画板—1")        # 创建 figure 实例 
ax_1 = fig_1.add_subplot(1, 1, 1)    # 创建 Axes 实例，具体参数设置请看下一章

ax_1.plot(X, C, label='cos()')
ax_1.plot(X, S)

ax_1.set_title('图像_1')                       # 显示图像标题，主要要在前面加个 "set_"
ax_1.set_xlabel('x 轴')                        # 显示图像 x轴 标题
ax_1.set_ylabel('y 轴')                        # 显示图像 y轴 标题
ax_1.set_xlim(-1, 1)                           # x 轴范围设置
ax_1.set_xticks([-0.9, -0.3, 0, -0.3])         # x 轴对应数字的替换
ax_1.grid(True)                                # 显示网格线
# 坐标轴的移动与之前一样
ax_1.legend(loc='lower right', title='图例')   # 让图例生效
ax_1.annotate('注释',                          # 显示注释
             xy=(0.5, 0.5),
             xytext=(1, 1),
             arrowprops = dict(facecolor = "k",
                               headlength = 5, 
                               headwidth = 5, 
                               width = 0.2)
            )


# 创建 画板-2 - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
fig_2 = plt.figure("画板—2")
ax_2 = fig_2.add_subplot(1, 1, 1)

ax_2.plot(X, C)
ax_2.plot(X, S)

ax_2.set_title('图像_2')              # 显示图像标题
ax_2.set_xlabel('x 轴')               # 显示图像 x轴 标题
ax_2.set_ylabel('y 轴')               # 显示图像 y轴 标题


# 创建 画板-3 - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
fig_3 = plt.figure("画板—3")
ax_3 = fig_3.add_subplot(1, 1, 1)

ax_3.plot(X, C)
ax_3.plot(X, S)

ax_3.set_title('图像_3')              # 显示图像标题
ax_3.set_xlabel('x 轴')               # 显示图像 x轴 标题
ax_3.set_ylabel('y 轴')               # 显示图像 y轴 标题


plt.show()
```

## plt.figure() 的基本属性

### 参数：

* num : 默认为1, 图像的ID，如果没有定义图像名字，窗口名为：figure num
* figsize : 图像的长和宽（英寸）
* dpi : 分辨率（点/英寸）
* facecolor ：绘图区域的背景颜色
* edgecolor : 绘图区域边缘的颜色
* frameon : 是否绘制图像边缘，默认 True
