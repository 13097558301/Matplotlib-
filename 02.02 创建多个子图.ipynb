对于如何使用单个 figure 绘制多张子图，其重点在于：如何合理的确定子图在整个 figure 中的位置，对于这部分而言 Matplotlib 有多种方法来进行实现，但方法一旦多了，就有些杂乱，于是我总结了两种相对简单且大部分功能均可实现的方法。

```python
import matplotlib.pyplot as plt
import numpy as np
```

准备数据

```python
X = np.linspace(-np.pi, np.pi, 300)

C = np.cos(X)
S = np.sin(X)
```

## 方法一：通过add_subplot（）实现子图的创建及划分

* 先是通过plt.figure()创建图像（画板），之后再通过add_subplot()在图像上创建不同的区域。

* 对于 add_subplot()的参数，类似(3, 1, 1)就是将图像分为3行1列的区域，使用第1个区域。

* 具体了解请看下图：

```python
fig = plt.figure()

ax_1 = fig.add_subplot(1, 3, 1)

ax_1.plot(X, C)
ax_1.plot(X, S)


ax_2 = fig.add_subplot(3, 1, 3)

ax_2.plot(X, C)
ax_2.plot(X, S)

ax_3 = fig.add_subplot(3, 3, 2)

ax_3.plot(X, C)
ax_3.plot(X, S)

ax_4 = fig.add_subplot(3, 3, 3)

ax_4.plot(X, C)
ax_4.plot(X, S)

ax_5 = fig.add_subplot(3, 3, 5)

ax_5.plot(X, C)
ax_5.plot(X, S)

ax_6 = fig.add_subplot(3, 3, 6)

ax_6.plot(X, C)
ax_6.plot(X, S)


plt.show()
```

## 方法二：GridSpec

对于方法一来说，有些分块实现不了，于是这里提供方法二，网格布局的话可以更加灵活的控制占用多少空间。先是通过gridspec.GridSpec()创建区域，参数3，3的意思就是每行3个，每列3个，将图像分为3行3列的区域。

gs[x:d_x, y:d_y] ；(x, y) 代表矩阵起始位置，(d_x, d_y) 代表矩阵结束的位置；注意：x 代表行， y 代表列， 从 0 开始。

```python
import matplotlib.gridspec as gridspec
```

```python
fig = plt.figure()

gs = gridspec.GridSpec(3, 3)

ax_1 = fig.add_subplot(gs[0:1, 0:3])

ax_2 = fig.add_subplot(gs[1:3, 0:1])

ax_3 = fig.add_subplot(gs[1:3, 1:2])

ax_4 = fig.add_subplot(gs[1:3, 2:3])


plt.show()
```
