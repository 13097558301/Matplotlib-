## 设置 x轴、y轴 的位置

这时候你必须创建figure、subplot。实际上每幅图有四条脊柱（上下左右），为了将脊柱放在图的中间，我们必须将其中的两条（上和右）设置为无色，然后调整剩下的两条到合适的位置——数据空间的 0 点。

```python
import numpy as np
import matplotlib.pyplot as plt
```

准备数据

```python
X = np.linspace(-np.pi, np.pi, 300)

C = np.cos(X)
S = np.sin(X)
```

设置 x轴、y轴 及 绘图

```python
fig = plt.figure()               # 创建一个 figure
ax = fig.add_subplot(1, 1, 1)    # 添加一个子图

ax.spines['right'].set_color('none')    # 将右边的轴隐藏
ax.spines['top'].set_color('none')      # 将上边的轴隐藏

ax.xaxis.set_ticks_position('bottom')   # 将下边的轴设为 x轴
ax.yaxis.set_ticks_position('left')     # 将左边的轴设为 y轴

ax.spines['bottom'].set_position(('data',0))      # 将 x轴，移至 0
ax.spines['left'].set_position(('data', 0))       # 将 y轴，移至 0

ax.plot(X, C)
ax.plot(X, S)

plt.show()
```
